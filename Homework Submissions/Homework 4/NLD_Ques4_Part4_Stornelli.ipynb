```python
import numpy as np
import matplotlib.pyplot as plt

tau = 1

def energy_function(theta, v, tau):
    return 0.5*v**2-np.cos(theta)-tau*theta

def phase_portrait(theta, v, tau):
    dtheta = v
    dv = tau-np.sin(theta)
    return dtheta, dv

theta = np.linspace(-2*np.pi, 2*np.pi, 50)
v = np.linspace(-3, 3, 50)
Theta, v= np.meshgrid(theta, v)
Energy = energy_function(Theta, v, tau)

dTheta, dv = phase_portrait(Theta, v, tau)

magnitude = np.sqrt(dTheta**2 + dv**2)
dTheta /= magnitude
dv /= magnitude

fig = plt.figure(figsize=(14, 10))
ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(Theta, v, Energy, cmap='jet', alpha=0.8)

ax.set_title("3D Energy Surface")
ax.set_xlabel("$\dot{\\theta}$")
ax.set_ylabel("$\dot{v}$")
ax.set_zlabel("Energy $E$")
ax.view_init(elev=30, azim=120)  

ax2 = fig.add_subplot(122)
ax2.quiver(Theta, v, dTheta, dv, color="black")
ax2.set_title("x-y Projection")
ax2.set_xlabel("$\dot{\\theta}$")
ax2.set_ylabel("$\dot{v}$")
ax2.axhline(0, color="black", linestyle="--", linewidth=0.5)
ax2.axvline(0, color="black", linestyle="--", linewidth=0.5)
ax2.grid()

plt.tight_layout()
plt.show()
```
