```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib widget

ax = plt.figure().add_subplot(projection='3d')
x = np.array(np.linspace(-0.5 * np.pi,1.5 * np.pi,250))
y = np.array(np.linspace(-5,5,250))
X, Y = np.meshgrid(x,y)
Z = 0.4*Y*Y - np.cos(X) -1


# Plot the 3D surface
ax.plot_surface(X, Y, Z, edgecolor='Purple', lw=0.5, rstride=20, cstride=20,
                alpha=0.1)

ax.contour(X, Y, Z, zdir='z', offset=-1, cmap='coolwarm')

ax.set(xlim=(-1/5*np.pi, 3/2*np.pi), ylim=(-6, 6), zlim=(0, 13),
       xlabel=r'$\theta$', ylabel=r'$\omega$', zlabel=r'E($\theta$,$\omega$)')

plt.show()
```
