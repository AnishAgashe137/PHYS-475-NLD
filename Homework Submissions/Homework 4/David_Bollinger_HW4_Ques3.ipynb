```python
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt

def diffeq(t,val):
    x,y = val
    return [1*y, -.5*y-np.sin(x)]
xiv, yiv = [3-.9,3-.9,3-.9,3.9,3.9,3.9,3,3], [-.9,0,.9,-.9,0,.9,-.9,.9]
#other initial conditions here
for i in range(8):
    sol = solve_ivp(diffeq,
                  t_span=[0,300],
                  y0=[xiv[i],yiv[i]],
                  t_eval=np.linspace(0,30,1000))
    plt.plot(sol.y[0,:], sol.y[1,:],label='($x_0$,$Y_0$) = ('+str(xiv[i])+','+str(yiv[i])+')')

x=np.linspace(-2, 8, 40)
y=np.linspace(-5, 5, 40)

X, Y = np.meshgrid(x,y)

DX,DY=diffeq(0, [X,Y])

#this normalizes the vectors
N=np.sqrt(DX**2 + +DY**2)
N[N==0]=1 #eliminates divide by 0 errors
DX/=N
DY/=N

plt.quiver(X,Y,DX,DY)
plt.legend(loc='center left', bbox_to_anchor=(1,0.5))
plt.plot([0,0],[-8,8],'k')
plt.plot([-8,8],[0,0],'k')
plt.xlim(-2,8)
plt.ylim(-5,5)
```
