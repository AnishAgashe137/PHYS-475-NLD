```python
import numpy as np
import matplotlib.pyplot as plt

def damped_pendulum(theta, v, b):
    dtheta = v
    dv = -b*v-np.sin(theta)
    return dtheta, dv

theta = np.linspace(-2 * np.pi, 2 * np.pi, 50)
v = np.linspace(-3, 3, 50)
Theta, V = np.meshgrid(theta, v)

b = 1

dTheta, dV = damped_pendulum(Theta, V, b)

# Normalize vectors for plotting
magnitude = np.sqrt(dTheta**2 + dV**2)
dTheta /= magnitude
dV /= magnitude

plt.figure(figsize=(10, 6))
plt.quiver(Theta, V, dTheta, dV, color="black", angles="xy")
plt.title("Phase Portrait of Damped Pendulum")
plt.xlabel("$\dot{\\theta}$")
plt.ylabel("$\dot{v}$")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.grid()
plt.show()
```
