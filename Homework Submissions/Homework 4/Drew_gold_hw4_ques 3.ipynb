```python
import numpy as np
import matplotlib.pyplot as plt

# Define the system of equations
def pendulum_system(theta, omega, b):
    dtheta = omega
    domega = -b * omega - np.sin(theta)
    return dtheta, domega

# Create a grid for theta and omega
theta = np.linspace(-2 * np.pi, 2 * np.pi, 100)
omega = np.linspace(-2, 2, 100)
Theta, Omega = np.meshgrid(theta, omega)

# Damping factor
b = 0.5

# Compute the vector field
DTheta, DOmega = pendulum_system(Theta, Omega, b)

# Normalize the vectors for better visualization
magnitude = np.sqrt(DTheta**2 + DOmega**2)
DTheta /= magnitude
DOmega /= magnitude

# Plot the phase portrait
plt.figure(figsize=(10, 6))
plt.quiver(Theta, Omega, DTheta, DOmega, color='b', pivot='middle', alpha=0.75)
plt.title("Phase Portrait of the Damped Pendulum", fontsize=14)
plt.xlabel(r"$\theta$", fontsize=12)
plt.ylabel(r"$\dot{\theta}$", fontsize=12)
plt.axhline(0, color='k', linewidth=0.5, linestyle='--')
plt.axvline(0, color='k', linewidth=0.5, linestyle='--')

# Mark fixed points
stable_points = [0, 2 * np.pi, -2 * np.pi]
unstable_points = [np.pi, -np.pi]
plt.scatter(stable_points, [0] * len(stable_points), color='g', label='Stable Fixed Points')
plt.scatter(unstable_points, [0] * len(unstable_points), color='r', label='Unstable Fixed Points')

plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
