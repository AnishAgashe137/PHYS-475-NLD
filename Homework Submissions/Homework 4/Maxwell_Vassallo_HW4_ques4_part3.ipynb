```python
import numpy as np
import matplotlib.pyplot as plt


def torque_pendulum(theta, x, tau):
    dtheta = x
    dx = tau - np.sin(theta)
    return dtheta, dx


theta = np.linspace(-2 * np.pi, 2 * np.pi, 30)
x = np.linspace(-2, 2, 30)
Theta, X = np.meshgrid(theta, x)

# Constants
tau = 0.5  

dTheta, dX = torque_pendulum(Theta, X, tau)



# Ploting the portrait
plt.figure(figsize=(10, 6))
plt.quiver(Theta, X, dTheta, dX, color='black', pivot='middle', alpha=0.75)
plt.xlabel(r"$\theta$", fontsize=12)
plt.ylabel(r"$\dot{\theta}$", fontsize=12)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
