```python
import numpy as np
import matplotlib.pyplot as plt

#constants / Inital Values
b = 1
#setting up equations for penedulum
def pendulum(theta, x, b): #b is standin for tau
    dtheta = x
    dx = b-np.sin(theta)
    return dtheta, dx

theta = np.linspace(-2 * np.pi, 2 * np.pi, 70) #creating the boundaries for the period
x = np.linspace(-5, 5, 30)
Theta, X = np.meshgrid(theta, x)
dTheta, dX = pendulum(Theta, X, b)

# Ploting the graph
plt.figure(figsize=(10, 5))
plt.quiver(Theta, X, dTheta, dX, color='Blue', pivot='middle', alpha=1)
plt.xlabel(r"$\theta$", fontsize=10)
plt.ylabel(r"$\dot{\theta}$", fontsize=12)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
```
