```python
import numpy as np
import matplotlib.pyplot as plt

# In the defined function tau can be set to any value 
def phase_portrait(theta, v, tau=1):
    dtheta = v
    dv = tau - np.sin(theta)
    return dtheta, dv


theta = np.linspace(-2 * np.pi, 2 * np.pi, 50)  
v = np.linspace(-2, 2, 50)
Theta, v = np.meshgrid(theta, v)

dTheta, dv = phase_portrait(Theta, v)

magnitude = np.sqrt(dTheta**2 + dv**2)
dTheta /= magnitude
dv /= magnitude


plt.figure(figsize=(10, 6))
plt.quiver(Theta, v, dTheta, dv, color="black", angles="xy")
plt.title("Phase Portrait of $\ddot{\\theta} + \sin(\\theta) = \\tau$")
plt.xlabel("$\dot{\\theta}$")
plt.ylabel("$\dot{v}$")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.grid()
plt.show()
```
