```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
g = 9.81  # acceleration due to gravity (m/s^2)
L = 1.0   # length of the pendulum (m)

# Define the system of equations
def pendulum_ode(theta, omega):
    dtheta_dt = omega
    domega_dt = -(g / L) * np.sin(theta)
    return dtheta_dt, domega_dt

# Grid for  (theta) and  (omega)
theta = np.linspace(-3 * np.pi, 3 * np.pi, 50)
omega = np.linspace(-10, 10, 50)
Theta, Omega = np.meshgrid(theta, omega)

# Calculate derivatives on the grid
dTheta, dOmega = pendulum_ode(Theta, Omega)

# Plotting the quiver plot
plt.figure(figsize=(10, 8))
plt.quiver(Theta, Omega, dTheta, dOmega, color='blue')

# Labeling the plot
plt.title('Phase Portrait of the Non-linear Pendulum')
plt.xlabel(r'$\theta$ (rad)')
plt.ylabel(r'$\omega$ (rad/s)')
plt.xlim([-3 * np.pi, 3 * np.pi])
plt.ylim([-10, 10])

# Show the plot
plt.grid(True)
plt.show()
```
