```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

r = [-10, -1, 0, 0.5, 1]

# function that returns dx/dt
def model(x,t,r):
    dxdt = r*x - np.sin(x)
    return dxdt

# initial condition
x0 = 3*np.pi/4

# time points
t = np.linspace(0,5)

for i in r:
    # solve ODE
    x = odeint(model,x0,t,args=(i,))

    # plot results
    plt.plot(t,x, linewidth = 1, color = 'hotpink')
    plt.title(["r = ", i,])
    plt.xlabel('time')
    plt.ylabel('x(t)')
    plt.show()
```
