```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

r=0
R=[-1,-2/(3*np.pi),0,2/(5*np.pi),1]

def fun(x):
    dxdt = (r*x)-np.sin(x)
    return dxdt

for num in R:
    r=num
    x = np.array(np.linspace(-10,10,100))
    X=fun(x)
    plt.plot(x,X,label='$r$ = {}'.format(num))
plt.xlabel(r'$x$')
plt.ylabel(r'$\dot{x}$')
plt.grid()
plt.legend()
plt.show()
```
