```python
import mpmath
import numpy as np
import matplotlib.pyplot as plt

def model(t,x):
    return t * x - mpmath.sin(x)
    

X0 = [-1, 0, .3, 3]
T0 = np.linspace(0,1e2)


for x in X0:
    y = [float(model(t,x)) for t in T0]
    plt.plot(T0, y, label=f'x = {x}')

#x = odeint(model,X0,t)

plt.xlabel('time')
plt.ylabel('r(t)')
plt.legend()
plt.show()
```
