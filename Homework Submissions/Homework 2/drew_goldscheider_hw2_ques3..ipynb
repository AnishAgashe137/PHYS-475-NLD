```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function for the 1D flow
def flow(x, r):
    return r * x - np.sin(x)

# Generate values of x
x_vals = np.linspace(-10, 10, 400)

# Values of r to consider
r_values = [-1, 0, 0.5, 1.5]

# Create subplots
fig, axes = plt.subplots(len(r_values), 1, figsize=(8, 6), sharex=True)

# Plot the flow for each value of r
for ax, r in zip(axes, r_values):
    f_vals = flow(x_vals, r)
    ax.plot(x_vals, f_vals, label=f'r = {r}')
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)
    ax.set_title(f'Flow of the system for r = {r}')
    ax.set_ylabel('$\\dot{x}$')
    ax.grid(True)

# Add xlabel to the last subplot
axes[-1].set_xlabel('x')

# Show the plot
plt.tight_layout()
plt.show()
```
