```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
#Function for 1D Flow
def Function(x,r):
    Equation = r * x - np.sin(x)
    return Equation

#Values for x
x_values = np.linspace(-10,10,400)

#Static R values for flow (Intial Conditions)
r_values = [-1, 0, 0.5, 2]

#labels for r matrix
labels= ['r=-1', 'r=0','r=0.5','r=2']
#Plotting Function for each value of r
plt.figure(figsize=(10,10))

for r, label in zip(r_values, labels):
    vals = Function(x_values,r)
    lab = r
    plt.plot(x_values,vals)
    plt.title(['r= ', r,])
    plt.show()

```
