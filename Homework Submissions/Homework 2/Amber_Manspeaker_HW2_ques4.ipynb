```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

l = 1

# function that returns dx/dt
def model(p,t,k):
    dpdt = l*np.sin(p)*(np.cos(p) - k)
    return dpdt

# initial condition
p0 = 0.1

# time points
t = np.linspace(0,10)

for h in np.linspace(0, 2):

# solve ODE
    p = odeint(model,p0,t, args=(h,))
    
    plt.plot(t,p,h)
    plt.xlabel('time')
    plt.ylabel('x(t)')
    plt.show()

#I am having trouble making the animation do what I want it to. This is the general idea of the goal of my animation. 
#I wanted to show how that x(t) line progresses 
```
