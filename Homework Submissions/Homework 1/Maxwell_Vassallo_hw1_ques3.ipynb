```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# time points
t = np.linspace(0,25)

#Define Constants g, k, m
g= 9.81
k= 5
m= 15

#Initial Velocity
V0 = 0

# function that returns dv/dt
def model(v,t):
    dvdt = g - ((k * v**2) / m)
    return dvdt

#ODE Solver
vdot = odeint(model, V0, t)

#Plotting Magic v(t) vs v
plt.plot(vdot,t)
plt.xlabel('Time(t)')
plt.ylabel('V(t)')
plt.title('V(t) vs v')
```

```python
#vdot vs v
dvdt = (v[1:]-v[:-1])/(t[1:]-t[:-1])
plt.plot(v[1:],dvdt)
plt.xlabel(r'$v$')
plt.ylabel(r'$\dot{v}$')
plt.show()

```
