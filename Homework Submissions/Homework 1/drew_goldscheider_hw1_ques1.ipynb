```python
import numpy as np
import matplotlib.pyplot as plt

# Define the inverted function x(t)
def x_t(t, x0):
    return 2 * np.arctan(np.exp(t) / (1/np.sin(x0) + np.cos(x0)/np.sin(x0)))

# Time values for plotting
t_vals = np.linspace(0, 5, 500)

# Initial conditions
x0_vals = [np.pi/4, np.pi/2, 7*np.pi/4]

# Plotting the curves for the three initial conditions
plt.figure(figsize=(8, 6))

for x0 in x0_vals:
    x_vals = x_t(t_vals, x0)
    plt.plot(t_vals, x_vals, label=f'$x_0 = {x0:.2f}$')

# Adding labels, title, and legend
plt.xlabel('Time $t$')
plt.ylabel('$x(t)$')
plt.title('Plot of $x(t) = 2\\tan^{-1}\\left(\\frac{e^t}{\\csc(x_0) + \\cot(x_0)}\\right)$')
plt.legend()
plt.grid(True)
plt.show()
```
