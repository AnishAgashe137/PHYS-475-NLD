```python
import mpmath
import numpy as np
import matplotlib.pyplot as plt

def model(t,x):
    return 2 * mpmath.atan(mpmath.exp(t) / (mpmath.csc(x) + mpmath.cot(x)))
    

X0 = [mpmath.pi/4, mpmath.pi/2, 7*mpmath.pi/4]
T0 = np.linspace(0,1e2)


for x in X0:
    y = [float(model(t,x)) for t in T0]
    plt.plot(T0, y, label=f'x = {x}')

#x = odeint(model,X0,t)

plt.xlabel('time')
plt.ylabel('x(t)')
plt.legend()
plt.show()
```
