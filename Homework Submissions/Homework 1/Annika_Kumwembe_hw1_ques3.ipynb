```python
import mpmath
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt

g = 9.82
k = 2.1
m = 25

def vtv(v,t):
    dvdt = g - (k*v*v)/m
    return dvdt

v0 = 0

t = np.linspace(0,20)

v = odeint(vtv, v0, t)

plt.plot(v,t)
plt.xlabel('time')
plt.ylabel('v(t)')
plt.show
```
