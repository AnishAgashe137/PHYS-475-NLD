```python
from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt

g = 10
l = 10


def diffeq(t,val):
  x,y = val
  return [(g/l)*np.sin(x), (g/l)*np.cos(x)]

#xiv, yiv = [-0.43,-2,2,2], [2,-2,-2,2]
x0 = [0.00001,-1,-.25,0.25,0.5,1,1.5,2,2.5,3]

for i in range(6):
    sol = solve_ivp(diffeq, 
                    t_span = [-3*np.pi,3*np.pi],
                    y0 = [x0[i],x0[i]],
                    t_eval=np.linspace(-3*np.pi,3*np.pi,100))
    plt.plot(sol.y[0,:], sol.y[1,:])
    #plt.plot(sol.t, sol.y[0],'-')
    #plt.plot(sol.t, sol.y[1],'-')

x = np.linspace(-np.pi, np.pi, 15) #creating an array for x
#y = np.linspace(-2, 2, 15) #creating an array for y

X, Y = np.meshgrid(x, x) #creating a meshgrid over the x-y phase space

DX, DY = diffeq(0, [X, Y]) #assigning the xdot andnydot value at each grid point



plt.quiver(X, Y, DX, DY) #plots the unit vector arrows at eac grid point
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.plot([0,0],[-5,5],'k')
plt.plot([-5,5],[0,0],'k')
plt.xlim(-np.pi, np.pi)
plt.ylim(-np.pi,np.pi)
plt.xlabel('x')
plt.ylabel('y')
```
