```python
#Phase diagram when there is no human intervention and the forest is left to burn

from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt

a = 0
b = 3
c = 20
d = 3

def diffeq(t,val):
  x,y = val
  return [x*(a-b*y), -y*(c-d*x)]

#xiv, yiv = [-0.43,-2,2,2], [2,-2,-2,2]
#xiv, yiv = [0,0,50,50,75,100,25], [0,25,50,25,50,75,0]
#xiv, yiv = [30],[10]
xiv, yiv = [35,45,25,30,55], [15,25,5,10,35]

for i in range(5):
    sol = solve_ivp(diffeq, 
                    t_span = [0,10],
                    y0=[xiv[i],yiv[i]],
                    t_eval=np.linspace(0,10,10000))
    plt.plot(sol.y[0,:], sol.y[1,:],label='($x_0$,$y_0$) = ('+str(xiv[i])+','+str(yiv[i])+')')
    #plt.plot(sol.t, sol.y[0],'-')
    #plt.plot(sol.t, sol.y[1],'-')

x = np.linspace(-100, 100 ,25) #creating an array for x
y = np.linspace(-100, 100, 25) #creating an array for y

X, Y = np.meshgrid(x, y) #creating a meshgrid over the x-y phase space

DX, DY = diffeq(0, [X, Y]) #assigning the xdot andnydot value at each grid point

# This normalizes the arrows so they are all of unit length and just show the direction
N = np.sqrt(DX**2 + DY**2)
N[N==0] = 1 # eliminate /0 errors, it is sort of optional.
DX /= N
DY /= N

plt.quiver(X, Y, DX, DY) #plots the unit vector arrows at eac grid point
#plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title("Forest Fire Spread")
#plt.plot([0,0],[-25,25],'k')
#plt.plot([-25,25],[0,0],'k')
plt.xlim(0,50)
plt.ylim(0, 75)
plt.xlabel('Number of Live Trees')
plt.ylabel('Number of Fires')
```

```python
#Phase diagram when there is human intervention and the fire is being put out as it spreads. 

from scipy.integrate import solve_ivp
import numpy as np
import matplotlib.pyplot as plt

a = 0
b = 3
c = 20
d = 3
e = 25
#Extingish rate

def diffeq(t,val):
  x,y = val
  return [x*(a-(b*y)), -y*(c-d*x)-y*e]

#xiv, yiv = [-0.43,-2,2,2], [2,-2,-2,2]
#xiv, yiv = [15,50,10,30,25], [5,25,10,10,25]

xiv, yiv = [35,45,25,30,55], [15,25,5,10,35]

for i in range(5):
    sol = solve_ivp(diffeq, 
                    t_span = [0,10],
                    y0=[xiv[i],yiv[i]],
                    t_eval=np.linspace(0,10,10000))
    plt.plot(sol.y[0,:], sol.y[1,:],label='($x_0$,$y_0$) = ('+str(xiv[i])+','+str(yiv[i])+')')
    #plt.plot(sol.t, sol.y[0],'-')
    #plt.plot(sol.t, sol.y[1],'-')

x = np.linspace(-100, 100 ,25) #creating an array for x
y = np.linspace(-100, 100, 25) #creating an array for y

X, Y = np.meshgrid(x, y) #creating a meshgrid over the x-y phase space

DX, DY = diffeq(0, [X, Y]) #assigning the xdot andnydot value at each grid point

# This normalizes the arrows so they are all of unit length and just show the direction
N = np.sqrt(DX**2 + DY**2)
N[N==0] = 1 # eliminate /0 errors, it is sort of optional.
DX /= N
DY /= N

plt.quiver(X, Y, DX, DY) #plots the unit vector arrows at eac grid point
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.title("Forest Fire Spread with Human Intervention")
#plt.plot([0,0],[-25,25],'k')
#plt.plot([-25,25],[0,0],'k')
plt.xlim(0,50)
plt.ylim(0,75)
plt.xlabel('Number of Live Trees')
plt.ylabel('Number of Fires')
```
