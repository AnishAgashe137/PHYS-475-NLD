# RC Circuit

Numerical code for solving the RC circuit equation:
$$ \dot{Q} = \frac{V_0}{R} - \frac{Q}{RC} $$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# constant values
V0 = 15
R = 5
C = 0.5

# function that returns dx/dt
def charge(q,t):
    dqdt = V0/R - q/(R*C)
    return dqdt

# initial condition
q0 = 5

# time points
t = np.linspace(0,10)

# solve ODE
q = odeint(charge,q0,t)

# plot results
plt.plot(t,q)
plt.xlabel('time')
plt.ylabel('Q(t)')
plt.show()
```

Plotting $\dot{Q}$ vs. $Q$ numerically

```python
dqdt = (q[1:]-q[:-1])/(t[1:]-t[:-1])
plt.plot(q[1:],dqdt)
plt.xlabel(r'$Q$')
plt.ylabel(r'$\dot{Q}$')
plt.show()
```
