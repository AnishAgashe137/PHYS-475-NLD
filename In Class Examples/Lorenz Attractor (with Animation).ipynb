## Lorenz Atrractor

Lorenz equations are given by,
$$ \dot{x} = \sigma(y - x)$$
$$ \dot{y} = rx - y - xz $$
$$ \dot{z} = xy - bz $$
where, $\sigma, r, b$ are positive parameters. 

For $\sigma = 10$, $b = \frac{8}{3}$, and $r > 13$, this system starts exhibiting chaotic behaviour. That is, the system follows aperiodic trajectories and hence its state cannot be predicted. The curious property is though that the trajectories are confined in a finite region in the phase space. Hence even though the system is unpredictable, it looks as though it is approaching a fixed point (an attractor). Such an "attractor" is known as a **strange attractor**. The strange attractor that the Lorenz systems exhibit is also knon as **Lorenz attractor**.
