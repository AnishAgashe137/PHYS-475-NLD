## Code for Plotting the Energy Surface 

### Particle moving in a Double Well Potential

The double well potential is given by,
$$ V(x) = \frac{1}{4} x^4 - \frac{1}{2} x^2 $$

The equation of motion is given by,
$$ m\ddot{x} + \frac{{\rm d} V}{{\rm d} x} = 0 $$

Using this, we can derive the energy function, given by,
$$ E(x,y) = \frac{1}{2} y^2 + \frac{1}{4} x^4 - \frac{1}{2} x^2 $$

The phase portrait of this particle contains $E$-constant trajectories. That is, the trajectories are the projection of the energy surface on the $x$-$y$ plane. 

```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib widget

ax = plt.figure().add_subplot(projection='3d')
x = np.array(np.linspace(-1.5,1.5,100))
y = np.array(np.linspace(-1.5,1.5,100))
X, Y = np.meshgrid(x,y)
Z = 0.5*Y*Y - 0.5*X*X + 0.25*(X**4) + 1 
# the (+1) in the end raises the 3D surface so that the projections are visible clearly

# Plot the 3D surface
ax.plot_surface(X, Y, Z, edgecolor='royalblue', lw=0.5, rstride=20, cstride=20,
                alpha=0.1)

# Plot projections of the contours for each dimension.  By choosing offsets
# that match the appropriate axes limits, the projected contours will sit on
# the 'walls' of the graph.
ax.contour(X, Y, Z, zdir='z', offset=0, cmap='coolwarm')
#ax.contour(X, Y, Z, zdir='x', offset=-2, cmap='coolwarm')
#ax.contour(X, Y, Z, zdir='y', offset=2, cmap='coolwarm')

ax.set(xlim=(-1.5, 1.5), ylim=(-1.5, 1.5), zlim=(0, 2),
       xlabel='x', ylabel='y', zlabel='E(x,y)')

plt.show()
```

### Pendulum

The equation for motion for a pendulum is given by,
$$ \ddot{\theta} +  \sin\theta = 0$$
where, the 'dot' represents a derivative with respect to, $\tau = \sqrt{\frac{g}{L}} t$.

Using this, the energy function is found as,
$$ E(\theta,\omega) = \frac{1}{2} \omega^2 - \cos\theta $$
where, $\omega = \dot{\theta}$.

Again, the phase portrait will be made of $E$-constant curves. 

```python
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib widget

ax = plt.figure().add_subplot(projection='3d')
x = np.array(np.linspace(-4*np.pi,4*np.pi,100))
y = np.array(np.linspace(-4,4,100))
X, Y = np.meshgrid(x,y)
Z = 0.5*Y*Y - np.cos(X) + 5
# the (+5) in the end raises the 3D surface so that the projections are visible clearly

# Plot the 3D surface
ax.plot_surface(X, Y, Z, edgecolor='royalblue', lw=0.5, rstride=20, cstride=20,
                alpha=0.1)

# Plot projections of the contours for each dimension.  By choosing offsets
# that match the appropriate axes limits, the projected contours will sit on
# the 'walls' of the graph.
ax.contour(X, Y, Z, zdir='z', offset=-1, cmap='coolwarm')
#ax.contour(X, Y, Z, zdir='x', offset=-2, cmap='coolwarm')
#ax.contour(X, Y, Z, zdir='y', offset=2, cmap='coolwarm')

ax.set(xlim=(-4*np.pi, 4*np.pi), ylim=(-6, 6), zlim=(0, 13),
       xlabel=r'$\theta$', ylabel=r'$\omega$', zlabel=r'E($\theta$,$\omega$)')

plt.show()
```
