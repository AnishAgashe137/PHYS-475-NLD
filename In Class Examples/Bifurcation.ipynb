# Bifurcations

## Saddle-Node Bifurcation

A saddle-node bifurcation has a normal form,
$$ \dot{x} = r + x^2 $$

**Example:**

Terminal velocity on various planets
$$ \dot{v} = g - \frac{k}{m} v^2$$


```python
import numpy as np
import matplotlib.pyplot as plt

# create a figure and axes
fig = plt.figure(figsize=(12,5))
ax1 = plt.subplot(1,2,1)

plt.rcParams['font.family'] = 'serif'

# set up the subplots as needed
ax1.set_xlim(( -15, 15))            
ax1.set_ylim((-80, 250))
ax1.set_xlabel('x')
ax1.set_ylabel('$\dot{x}$')

# create objects that will change in the animation. These are
# initially empty, and will be given new values for each frame
# in the animation.
txt_title = ax1.set_title('')
line1, = ax1.plot([], [], 'b', ls='-', lw=2)     # ax.plot returns a list of 2D line objects
line2, = ax1.plot([], [], 'k', ls='--', lw=2)
line3, = ax1.plot([], [], 'k', ls='--', lw=2)
```

```python
x = np.linspace(-15, 15, 1000)
r = np.linspace(-20,20,20)

# animation function. This is called sequentially
def drawframe(n):
    xdot = r[n] + x*x
    line1.set_data(x, xdot)
    line2.set_data([-15,15],[0,0])
    line3.set_data([0,0],[-100,500])
    txt_title.set_text('r = {0:4f}'.format(r[n]))
    return (line1,line2,line3)
```

```python
from matplotlib import animation

frame_num = len(r)  #sets the number of frames to be the length of the array for 'r' 
anim = animation.FuncAnimation(fig, drawframe, frames=frame_num, interval=500)#blit=True) 
                                                                              #blit=True re-draws only the parts
                                                                              #that have changed.
```

```python
from IPython.display import HTML #only necessary to display the video in-line in jupyer notebook
HTML(anim.to_html5_video())
```

## Transcritical Bifurcation
**Normal form**:
$$ \dot{x} = rx + x^2 $$

**Example:**

Laser Threshold
$$\dot{n} = (GN_0 - k)n - \alpha G n^2$$

```python
# create a figure and axes
fig = plt.figure(figsize=(12,5))
ax1 = plt.subplot(1,2,1)

plt.rcParams['font.family'] = 'serif'

# set up the subplots as needed
ax1.set_xlim(( -20, 20))            
ax1.set_ylim((-120, 250))
ax1.set_xlabel('x')
ax1.set_ylabel('$\dot{x}$')

# create objects that will change in the animation. These are
# initially empty, and will be given new values for each frame
# in the animation.
txt_title = ax1.set_title('')
line1, = ax1.plot([], [], 'b', ls='-', lw=2)     # ax.plot returns a list of 2D line objects
line2, = ax1.plot([], [], 'k', ls='--', lw=2)
line3, = ax1.plot([], [], 'k', ls='--', lw=2)

#ax1.legend(['sin','cos']);

x = np.linspace(-15, 15, 1000)
r = np.linspace(-20,20,20)

# animation function. This is called sequentially
def drawframe(n):
    xdot = r[n]*x + x*x
    line1.set_data(x, xdot)
    line2.set_data([-30,30],[0,0])
    line3.set_data([0,0],[-120,500])
    txt_title.set_text('r = {0:4f}'.format(r[n]))
    #ax1.set_ylim(min(xdot),max(xdot))
    return (line1,line2,line3)

frame_num = len(r)
anim = animation.FuncAnimation(fig, drawframe, frames=frame_num, interval=1000)#, blit=True)
```

```python
HTML(anim.to_html5_video())
```
