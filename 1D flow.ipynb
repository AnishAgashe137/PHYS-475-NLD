Solving the equation for an oscillatory flow (in one dimension)
$$ \dot{x} = \sin x $$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dy/dt
def model(x,t):
    dxdt = np.sin(x)
    return dxdt

# initial condition
x0 = np.pi/4.

# time points
t = np.linspace(0,5)

# solve ODE
x = odeint(model,x0,t)

# plot results
plt.plot(t,x)
plt.xlabel('time')
plt.ylabel('x(t)')
plt.show()
```

```python
dxdt = (x[1:]-x[:-1])/(t[1:]-t[:-1])
plt.plot(x[1:],dxdt)
plt.xlabel(r'$x$')
plt.ylabel(r'$\dot{x}$')
plt.show()
```
